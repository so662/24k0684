```python
class SecurityAgent:
    def __init__(self, agent_id, name, status):
        self.agent_id = agent_id
        self.name = name
        self.status = status

class FirewallAgent(SecurityAgent):
    def monitor_traffic(self):
        print(f"{self.name} is monitoring network traffic")

class MalwareDetectionAgent(SecurityAgent):
    def scan_files(self):
        print(f"{self.name} is scanning files for threats")

class AutomationAgent(SecurityAgent):
    def run_automation(self):
        print(f"{self.name} is performing AI-based tasks")

firewall = FirewallAgent("FW001", "FirewallAgent", "Active")
malware = MalwareDetectionAgent("MD002", "MalwareDetectionAgent", "Active")
automation = AutomationAgent("AA003", "AutomationAgent", "Active")

firewall.monitor_traffic()
malware.scan_files()
automation.run_automation()
```

```python
class ThreatIntelligence:
    def __init__(self, threat_id, name, severity):
        self.threat_id = threat_id
        self.name = name
        self.severity = severity

class PhishingThreat(ThreatIntelligence):
    def analyze_email(self):
        print(f"{self.name} is analyzing emails")

class RansomwareThreat(ThreatIntelligence):
    def scan_files(self):
        print(f"{self.name} is scanning file systems")

class BotnetThreat(ThreatIntelligence):
    def detect_traffic(self):
        print(f"{self.name} is detecting network traffic")

phishing = PhishingThreat("T001", "PhishingThreat", "High")
ransomware = RansomwareThreat("T002", "RansomwareThreat", "Critical")
botnet = BotnetThreat("T003", "BotnetThreat", "Medium")

phishing.analyze_email()
ransomware.scan_files()
botnet.detect_traffic()
```

```python
class ResponseAgent:
    def execute_response(self):
        print("Base response")

class AlertAgent(ResponseAgent):
    def execute_response(self):
        print("AlertAgent is sending notifications")

class BlockAgent(ResponseAgent):
    def execute_response(self):
        print("BlockAgent is blocking malicious activities")

class RecoverAgent(ResponseAgent):
    def execute_response(self):
        print("RecoverAgent is restoring affected systems")

alert = AlertAgent()
block = BlockAgent()
recover = RecoverAgent()

agents = [alert, block, recover]

for agent in agents:
    agent.execute_response()
```

```python
class Employee:
    def work(self):
        print("Employee is working")

class Manager(Employee):
    def work(self):
        print("Manager is managing")

class Developer(Employee):
    def work(self):
        print("Developer is coding")

class Designer(Employee):
    def work(self):
        print("Designer is designing")

manager = Manager()
developer = Developer()
designer = Designer()

manager.work()
developer.work()
designer.work()
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.__grade = None
    
    def set_grade(self, grade):
        self.__grade = grade
    
    def get_grade(self):
        return self.__grade
    
    def display_info(self):
        print(f"Student: {self.name}, Grade: {self.get_grade()}")

student1 = Student("AHMED")
student2 = Student("ALI")

student1.set_grade(85)
student2.set_grade(92)

student1.display_info()
student2.display_info()
```
